```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
rgb=pd.read_csv('rgb_data.csv',names=['First','Second','Third'])
rgb.name='Rgb color tranceiver'
receiver=pd.read_csv('receiver_data.csv',names=['First','Second','Third'])
receiver.name='Color recognition'
```

```python
def counts_statistics(dt):
    data={'red':dt.value_counts().red,
          'green':dt.value_counts().green,
          'bleu':dt.value_counts().bleu}
    values=data.values()
    colors=data.keys()
    return values,colors,dt.name
def make_plots(items):
    fig=plt.figure(figsize=(3,3))
    plt.bar(items[1],items[0],color=['red','green','blue'],width=0.4)
    plt.title(items[2])
    plt.xlabel('Colors')
    plt.ylabel('Frequency')
```

```python
print("Rgb colors")
make_plots(counts_statistics(rgb.First))
print("Tranceiver's recognition")
make_plots(counts_statistics(receiver.First))
```

```python
print("Rgb colors")
make_plots(counts_statistics(rgb.Second))
print("Tranceiver's recognition")
make_plots(counts_statistics(receiver.Second))
```

```python
print("Rgb colors")
make_plots(counts_statistics(rgb.Third))
print("Tranceiver's recognition")
make_plots(counts_statistics(receiver.Third))
```

```python
compare_df=rgb.compare(receiver,keep_shape=True)
```

```python
first_corrects=50-(compare_df.First.self.value_counts().red+compare_df.First.self.value_counts().bleu+compare_df.First.self.value_counts().green)
second_corrects=50-(compare_df.Second.self.value_counts().red+compare_df.Second.self.value_counts().bleu+compare_df.Second.self.value_counts().green)
third_corrects=50-(compare_df.Third.self.value_counts().red+compare_df.Third.self.value_counts().bleu+compare_df.Third.self.value_counts().green)
```

```python
print('Total corrects at the first color:'+str(first_corrects))
print('Total corrects at the second color:'+str(second_corrects))
print('Total corrects at the third color:'+str(third_corrects))
```

```python
total_correct_guesses=first_corrects+second_corrects+third_corrects
accuracy=total_correct_guesses/150
print(str(float(accuracy)*100)+"%")
```
